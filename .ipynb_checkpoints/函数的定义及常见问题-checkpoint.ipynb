# 如何正确的定义函数*
<i> *指数学函数, 而不是python函数 </i>

### 1.定义一个Python函数
使用此方法定义的函数可用于绘图, 但无法求导/求积分

```python
def f(z): 
    return z^2
```

```python
#绘图
plot(f,0,2)
```

注意这里绘图的时候不可以用plot(f(z),0,2), 否则会发生NameError,因为z是f内部的变量

如果将z定义为一个variable, 就可用f(z)了

```python
var('z')   # define z to be a variable
f(z)
```

此时f(z)成为了一个'符号表达式'

```python
plot(f(z), 0, 2)
```

### 2.定义一个符号表达式
可以被调用,可以绘图,也可以求导数和积分

```python
g(x) = x^2
g
```

```python
#求导数
Dg = g.derivative()
Dg
```

```python
type(g)
```

```python
plot(g,0,2)
```

请注意,g和g(x)是两个不同的表达式, 尽管它们的形式和作用都很相似.

第5条说明了它们的区别

# 以下是关于函数的常见问题
### 4.意外的求值
用方法1定义一个分段函数:

$h(x)=\begin{cases}
0(x<2) \\
x-2(x\ge2)
\end{cases}$

```python
def h(x):
    if x<2:
        return 0
    else:
        return x-2
```

之后尝试对h(x)绘图:

```python
plot(h(x),0,4)
```

出现了明显的错误:这不是分段函数h的图像,而是直线y=x-2

错误的原因是: 在执行plot(h(x),0,4)时, h(x)会被求值, 得到表达式x-2

```python
h(x)
```

这是因为x被当作了Sage中的变量, 在判断if条件x<2时, 被求值并转化为了False

```python
x < 2
```

```python
bool(x<2)
```

正确的绘图方法是对h绘图,而非h(x)

```python
plot(h,0,4)
```

### 5.定义函数时错误地产生了常数
尝试定义 $f(x)=x$ 并对其求其导函数$g(x)$

```python
f = x
g = f.derivative()
g
```

结果看起来是正确的, 但当我们尝试调用'函数'g时就会发生错误

```python
g(3)   #error
```

发生错误的原因是: g是表达式'1', 其中不包含任何变量, 是一个常数, 无法接受参数3

有3种方法可以使g成为函数

```python
f(x) = x        #修改f的定义
g = f.derivative()
g
```

```python
f = x
g(x) = f.derivative()  #修改g的定义
g
```

```python
f = x
g = f.derivative()
g(x=3)        #显式的指定一个变量
```

可以使用variables和arguments方法来区分常数和函数:

```python
f(x) = x
g(x) = f.derivative()    #函数
print(g.variables()) # 出现在g中的变量
print(g.arguments())  # g可以接受的变量
```

```python
f = x
h = f.derivative()   #常数
print(h.variables()) # 出现在g中的变量
print(h.arguments())  # g可以接受的变量
```
